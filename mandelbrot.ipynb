```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
%matplotlib widget
```

```python
xmin = -2.5
xmax = 1.5
ymin = -1.5
ymax = 1.5
maxit = 5000
nx = 1024
ny = 768
x = np.linspace(xmin, xmax, nx)
y = np.linspace(ymin, ymax, ny)
X, Y = np.meshgrid(x, y)
Pixel = X + 1j*Y
```

```python
Z = np.zeros_like(Pixel)

mask = np.ones_like(Pixel, dtype=bool)
iters = np.zeros_like(Pixel, dtype=int)

for i in tqdm(range(maxit)):
    Z[mask] = Z[mask]**2 + Pixel[mask]
    mask = Z*Z.conj() < 16
    iters[mask] = i+1
    #Z = Z*mask  
```

```python
plt.figure()
plt.imshow(np.log(iters), extent=(xmin, xmax, ymin, ymax), cmap='viridis')
plt.show()
```
